```python
import os
import os.path as op
import nipype.pipeline.engine as pe             # pypeline engine
import pyacwereg.workflows.epi as epi
```

```python
# set up folders
subject_id = 'S400'
data_dir = op.abspath( '/home/annalisa/data' )
freesurfer_subjects_dir = op.join( data_dir, 'IXI', 'FREESURFER' )
subject_path = op.join( data_dir, 'IXI', 'subjects', subject_id )
```

```python
prep = epi.smri_preparation()
prep.base_dir = op.abspath( op.join( data_dir, 'results' ) )

prep.inputs.inputnode.subject_id = subject_id
prep.inputs.inputnode.fs_subjects_dir = freesurfer_subjects_dir
prep.inputs.inputnode.in_fmap_mag = op.join( data_dir, 'DTI', 'S000', 'ts0', 'FM_mag.nii.gz' )
prep.inputs.inputnode.in_fmap_pha = op.join( data_dir, 'DTI', 'S000', 'ts0', 'FM_pha.nii.gz' )
prep.inputs.inputnode.in_t1w_brain = op.join( freesurfer_subjects_dir, subject_id, 'mri', 'brain.finalsurfs.mgz' )
prep.inputs.inputnode.in_t2w = op.join( subject_path, 'T2-Guys-0977.nii.gz' )
```

```python
from IPython.display import Image
prep.write_graph()
Image( filename=op.join( prep.base_dir, prep.name, 'graph.dot.png' ) )
```

```python
prep.run()
```

```python
dist = epi.distortion_workflow()
dist.base_dir = op.abspath( op.join( data_dir, 'results' ) )
```

```python
dist.inputs.inputnode.in_file = op.join( prep.base_dir, prep.name, 'Combine', 'brain.finalsurfs_out_corrected_merged.nii.gz' )
dist.inputs.inputnode.in_pha = op.join( prep.base_dir, prep.name, 'CheckPhaseMap', 'FM_pha_out_trans_checked.nii.gz' )
dist.inputs.inputnode.in_mag = op.join( prep.base_dir, prep.name, 'Brain_FMap_Mag', 'transform_Warped_masked.nii.gz' )
dist.inputs.inputnode.in_mask = op.join( prep.base_dir, prep.name, 'OutMask', 'transform_Warped_masked_thresh.nii.gz' )
dist.inputs.inputnode.in_tpms = [ op.join( prep.base_dir, prep.name, 'SegmentT1', 'brain.finalsurfs_out_corrected_pve_%d.nii.gz' ) % r for r in range(3) ]
dist.inputs.inputnode.te_incr = 2.46e-3
dist.inputs.inputnode.echospacing = 0.77e-3/(3.0*128)
dist.inputs.inputnode.enc_dir = 'y-'
```

```python
dist.write_graph()
Image( filename=op.join( dist.base_dir, dist.name, 'graph.dot.png' ) )
```

```python
res_dist = dist.run()
```
