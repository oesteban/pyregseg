```python
import os
import os.path as op
import nipype.pipeline.engine as pe             # pypeline engine
import pyacwereg.workflows.epi as epi
```

```python
# set up folders
subject_id = 'S400'
data_dir = op.abspath( '/home/annalisa/data' )
freesurfer_subjects_dir = op.join( data_dir, 'IXI', 'FREESURFER' )
subject_path = op.join( data_dir, 'IXI', 'subjects', subject_id )
```

```python
prep = epi.smri_preparation()
prep.base_dir = op.abspath( op.join( data_dir, 'results' ) )

prep.inputs.inputnode.subject_id = subject_id
prep.inputs.inputnode.fs_subjects_dir = freesurfer_subjects_dir
prep.inputs.inputnode.in_fmap_mag = op.join( data_dir, 'DTI', 'S000', 'ts0', 'FM_mag.nii.gz' )
prep.inputs.inputnode.in_fmap_pha = op.join( data_dir, 'DTI', 'S000', 'ts0', 'FM_pha.nii.gz' )
prep.inputs.inputnode.in_t1w_brain = op.join( freesurfer_subjects_dir, subject_id, 'mri', 'brain.finalsurfs.mgz' )
prep.inputs.inputnode.in_t2w = op.join( subject_path, 'T2-Guys-0977.nii.gz' )
```

```python
from IPython.display import Image
prep.write_graph()
Image( filename=op.join( prep.base_dir, prep.name, 'graph.dot.png' ) )
```

```python
prep.run()
```
