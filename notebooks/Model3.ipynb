## Generate Surfaces

```python
l = nib.load( os.path.join( model_path, 'model_labeling.nii.gz' ) )
spacing = (1.0,1.0,1.0)
orig = tuple( [l.get_header()['qoffset_x'],l.get_header()['qoffset_y'],l.get_header()['qoffset_z'] ])
dim = l.get_shape()
extent=(0,dim[0]-1,0,dim[1]-1,0,dim[2]-1)
```

```python
from tvtk.api import tvtk
def GenerateSurface( data, fname ):
    grid = tvtk.ImageData(spacing=spacing, origin=orig, dimensions=data.shape)
    grid.point_data.scalars = data.ravel().astype( np.uint8 )
    grid.point_data.scalars.name = 'scalars'
    
    w2 = tvtk.StructuredGridWriter( input=grid, file_name=os.path.join( model_path, 'image.vti' ) )
    w2.write()
    
    iso = tvtk.ImageMarchingCubes(input=grid)
    iso.update()
    vtkmeshclean = tvtk.CleanPolyData( input=iso.output )
    vtkmeshclean.update()
    smooth = tvtk.SmoothPolyDataFilter( input=vtkmeshclean.output, number_of_iterations=20 )
    smooth.update()
    vtkmeshclean2 = tvtk.CleanPolyData( input=smooth.output )
    vtkmeshclean2.update()
    contour = vtkmeshclean2.output
    #contour.point_data.scalars = np.ones( np.shape( contour.points )[0] )
    #contour.point_data.scalars.name = 'scalars'
    w = tvtk.PolyDataWriter(input=contour, file_name=os.path.join( model_path, fname ) )
    w.write()
    return contour
```

<h2>Deformation field generation</h2>

```python
from scipy.interpolate import griddata
from scipy.ndimage.interpolation import map_coordinates
from scipy.ndimage.filters import gaussian_filter
import matplotlib.pyplot as plt
```

```python
msk = nib.load('Model3/signal_mask.nii.gz')
img = nib.load('Model3/signal_b0_merged.nii.gz')
datashape = msk.get_shape()
x,y,z = np.mgrid[0:datashape[0],0:datashape[1], 0:datashape[2] ]
img_data = img.get_data()
msk_data = msk.get_data()
```

```python
z_slice = 60

field = np.zeros( shape=(datashape[0],datashape[1],datashape[2],3) )
field[ 60,180,z_slice] = (0, -10, 0)
field[ 90,190,z_slice] = (0, 11, 0)
field[120,180,z_slice] = (0, -10, 0)

#points = np.concatenate( ([ (50,50,33), (50,100,33), (50,150,33) ],np.random.randint( 0, 101, size=(100,3) ) ))
#values = np.zeros( shape=np.shape(points) )
#values[0:3]=[ (0, -10, 0), (0, 11, 0), (0, -10, 0) ]

#field = griddata(points, values, (x,y,z), method='linear', fill_value=0)
field[:,:,:,1] = gaussian_filter( field[:,:,:,1], sigma=15 )
minf = np.min( field[:,:,:,1] )
field[:,:,:,1]/= -minf
field[:,:,:,1]*= 6.0
```

```python
extent=[0,datashape[1],0,datashape[0]]
plt.imshow( field[:,:,z_slice,1], extent=extent )
plt.colorbar()
plt.imshow( msk_data[:,:,z_slice], extent=extent, alpha=0.4 )
```

```python
nii = nib.Nifti1Image( field, np.identity(4), img.get_header() )
nib.save( nii, 'Model3/deformed_field.nii.gz' )
```

```python
grid_points =  np.array( (x,y,z) ).T
new_grid_points = grid_points - np.swapaxes(field,0,2)
new_positions = new_grid_points.reshape(-1,3).T
```

```python
new_img = np.array( [ map_coordinates( img_data[:,:,:,i], new_positions ).reshape( datashape ) for i in range(0, img_data.shape[-1] ) ] )
new_img[new_img<0.0] = 0.0
```

```python
new_mask = map_coordinates( msk_data, new_positions ).reshape( datashape )
new_mask[new_mask<=0.5] = 0.0
new_mask[new_mask>0.0] = 1.0
```

```python
niiHS = nib.Nifti1Image( new_img.T, img.get_affine(), img.get_header() )
nib.save( niiHS, 'Model3/deformed_signal.nii.gz' )
niiHS = nib.Nifti1Image( new_mask.T, msk.get_affine(), msk.get_header() )
nib.save( niiHS, 'Model3/deformed_mask.nii.gz' )
```

```python
import nipype.interfaces.base as base        # base pipeline interface (e.g. command line execution)
import nipype.interfaces.io as nio           # Data i/o
import nipype.interfaces.utility as util     # utility
import nipype.algorithms.misc as misc
import nipype.pipeline.engine as pe          # pypeline engine
import nipype.interfaces.fsl as fsl          # fsl
import nipype.interfaces.freesurfer as fs    # freesurfer
import nipype.interfaces.diffusion_toolkit as dtk
import nipype.interfaces.mrtrix as mrtrix   #<---- The important new part!
import nipype.interfaces.camino as camino

fsl.FSLCommand.set_default_output_type('NIFTI_GZ')
```

```python
base_dir = os.path.abspath( os.path.abspath( 'Model3' ) )
pipeline = pe.Workflow(name="Model_Process", base_dir = base_dir)

fsmask = pe.Node(interface=fs.ApplyMask(), name='MaskRawData' )
fsmask.inputs.in_file = os.path.abspath( os.path.join( 'Model3', 'deformed_signal.nii.gz' ))
fsmask.inputs.mask_file = os.path.abspath( os.path.join( 'Model3', 'deformed_mask.nii.gz' ))

datasink = pe.Node(nio.DataSink(base_directory=base_dir,parameterization=False), name='sinker')
datasink.container = 'deformed'

dtifit = pe.Node( interface=fsl.DTIFit(output_type='NIFTI_GZ'), name='TensorFitting' )
dtifit.inputs.bvecs = os.path.abspath( os.path.join( 'Model3', 'signal.bvec' ) )
dtifit.inputs.bvals = os.path.abspath( os.path.join( 'Model3', 'signal.bval' ) )
dtifit.inputs.mask = os.path.abspath( os.path.join( 'Model3', 'deformed_mask.nii.gz' ) )
dtifit.inputs.dwi = os.path.abspath( os.path.join( 'Model3', 'deformed_signal.nii.gz' ) )

# connect nodes
pipeline.connect([
                        (dtifit,datasink, [( 'FA','@FA'), ('MD','@ADC') ] )
                      ])
```

```python
import math
def get_phi( x, y ):
    if x==0 and y==0:
        return 0
    elif x>=0:
        return math.asin( y / math.sqrt( x**2 + y**2 ) )
    else:
        return -math.asin( y / math.sqrt( x**2 + y**2 ) ) + math.pi
        
vphi = np.vectorize( get_phi )
```

```python
from tvtk.api import tvtk
import matplotlib.mlab as ml
import nibabel as nib

def plotSlice( sliceId, points1, points2, fa_data, extent ):
    plt.imshow( fa_data[sliceId,:,:], extent=extent, cmap=cm.gray )
    plt.xticks( [] )
    plt.yticks( [] )
    
    cont1 = points1[ points1[:,1] == sliceId ]
    o = (0.5*(fa_data.shape[0]-1),0.5*(fa_data.shape[2]-1))
    phis = vphi( cont1[:,0]-o[0], cont1[:,2]-o[1])
    res = np.take( cont1, np.argsort(phis), axis=0 )
    plt.plot( res[:,0], res[:,2], linewidth=3 )
    
    if points2==None:
        return
    
    cont2 = points2[ abs(points2[:,1] - sliceId)<0.5 ]
    phis2 = vphi( cont2[:,0]-o[0], cont2[:,2]-o[1])
    res2 = np.take( cont2, np.argsort(phis2), axis=0 )
    plt.plot( res2[:,0], res2[:,2], linewidth=3, color='r' )
```
